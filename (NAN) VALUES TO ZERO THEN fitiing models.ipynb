```python
#Read and load hittters.csv on a panda data frame

# Also uncomment here and edit if you want to use then comment that other below
#hitters = pd.read_csv("hitters.csv", thousands=',')

# comment this
df = pd.read_csv("Hitters.csv", thousands=',')
df.head()
```

```python
#Inverstigate columNs available
df.columns
```

```python
#discard unneccesary columns
cols_to_use = ['Unnamed: 0', 'Runs',  'Walks', 'Years',
               'League','Division', 'PutOuts','Assists',
               'Errors', 'Salary', 'NewLeague']
df = df[cols_to_use]
df.head()
```

```python
#view new columns
df.shape
```

```python
#Investigate column salary
df.Salary
```

```python
#check for NAN values using isna function and there sum
#Our target has 59 NAN vaues
df.isna().sum()
```

```python
#handle this NAN values by adding a zero
# results shows no NA values anymore
col_to_fill_zero = ['Salary']
df[col_to_fill_zero] = df[col_to_fill_zero].fillna(0)
df.isna().sum()
```

```python
# we want now to remove files with string values (text)
# Results show data with no dummies

df =pd.get_dummies (df, drop_first=True)
df.head()

#column_to_reshape = Hitter['Salary'].values  # Extract the column as a NumPy array
#reshaped_column = column_to_reshape.reshape(-1, 1)  # Reshape the column

#Hitter['Salary'] = reshaped_column
#lets plot salary fro test as aour sample distribution
#X = Hitter.drop ('Salary', axis=1)
```

```python
#AFTER removed text items
df.shape
```

```python
# supress  the x axis

X = df.drop ('Salary', axis=1)
#X = hitters.drop ['Salary']
y = df[['Salary']]
```

```python
# Training specifics

#Usually this are basic standard methods and procedurers
#recognize we want to use 30% of data for tESTING, 70% FOR TRAINING
from sklearn.model_selection import train_test_split

train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.3, random_state=2)
```

```python
#Let us illustrate how to avoid overfitting through regularization
#  below code trains and fits the model

#lasso regression L1 regularization which adds an absolute theta value to the regularization model formular

from sklearn import  linear_model
#Lasso
Lasso_reg = linear_model.Lasso(alpha=50, max_iter=100, tol =0.1)
Lasso_reg.fit(train_X, train_y)

```

```python
# let us see the score for train data
#results not good though
Lasso_reg.score(train_X, train_y)
```

```python
# let us see the score for test data
Lasso_reg.score(test_X, test_y)
```

```python
# predicted target salary values if a penalty term is introduced to encourage the model uses less features in regression
# evaluating the perfomance by predicting salary items
pred_y = Lasso_reg.predict(test_X)
pred_y
```

```python
#ridge regression l2 regularization which adds a square theta value to the regularization model formular
from sklearn.linear_model import Ridge

Ridge_reg = Ridge(alpha=50, max_iter=100, tol =0.1)
Ridge_reg.fit(train_X, train_y)
```

```python
# let us see the score for train data
#results not good though BUT BETTER
Ridge_reg.score(train_X, train_y)
```

```python
# let us see the score for test data
Ridge_reg.score(test_X, test_y)
```

```python
# predicted target salary values if a penalty term is introduced to encourage the model uses less features in regression
# evaluating the perfomance by predicting salary items
pred_y = Ridge_reg.predict(test_X)
pred_y
```
