```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import sklearn #Simple and efficient tools for predictive data analysis
```

```python
#suppress warnings for clean notebook
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("Hitters.csv", thousands=',')
df.head()
```

```python
#discard unneccesary columns
cols_to_use = ['Unnamed: 0', 'Runs',  'Walks', 'Years',
               'League','Division', 'PutOuts','Assists',
               'Errors', 'Salary', 'NewLeague']
df = df[cols_to_use]
df.head()
```

```python

df =pd.get_dummies (df, drop_first=True)
df.head()
```

```python
#Investigate column salary
df.Salary
```

```python
# for our NAN cells let us try to use MEAN IN THIS CASE
# Find mean

mean = df['Salary'].mean()
print("Mean:", mean)
```

```python
#Fill the cells with NAN values with the mean
#y = y.fillna(y.mean())
df['Salary'].fillna(df['Salary'].mean(), inplace=True)
```

```python
X = df.drop ('Salary', axis=1)
#X = hitters.drop ['Salary']
y = df[['Salary']]
```

```python
# Training specifics

#Usually this are standard methods and procedurers
#recognize we want to use 20% or 30% of data for tESTING, 70% FOR TRAINING
from sklearn.model_selection import train_test_split

train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
#Let us illustrate how to avoid overfitting through regularization
#  below code trains and fits the model

#lasso regression L1 regularization which adds an absolute theta value to the regularization model formular

from sklearn import  linear_model
#Lasso
Lasso_reg = linear_model.Lasso(alpha=50, max_iter=100, tol =0.1)
Lasso_reg.fit(train_X, train_y)
```

```python
# let us see the score for train data
#when we used insa() function our test score was '0.29294827843996685' which means the mean made it worse
Lasso_reg.score(train_X, train_y)
```

```python
# let us see the score for test data
#while using insa() function our test score was '0.40230967364480186' which means the mean made it worse
Lasso_reg.score(test_X, test_y)
```

```python
pred_y = Lasso_reg.predict(test_X)
pred_y
```

```python
#ridge regression l2 regularization which adds a square theta value to the regularization model formular
from sklearn.linear_model import Ridge

Ridge_reg = Ridge(alpha=50, max_iter=100, tol =0.1)
Ridge_reg.fit(train_X, train_y)
```

```python
# let us see the score for train data
#when we added zeros in-place of NAN the score was 0.34477059384446285
Ridge_reg.score(train_X, train_y)
```

```python
# let us see the score for test data
#when we added zeros in-place of NAN the score was 0.41435000481178974

Ridge_reg.score(test_X, test_y)
```

```python
pred_y = Ridge_reg.predict(test_X)
pred_y
```
